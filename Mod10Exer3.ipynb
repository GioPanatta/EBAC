1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("previsao_de_renda.csv")

````

````python
data.head()
````

````python
# 1. Definindo a lista manual de variáveis qualitativas e atribuindo à variável "variable"
variable = ['sexo', 'posse_de_veiculo', 'posse_de_imovel', 'tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia']

# 2. Configurando o estilo e tamanho padrão do Seaborn
sns.set_style("whitegrid")
plt.figure(figsize=(16, 5))

# 3. Loop para plotar cada variável usando countplot
for var in variable:
    plt.figure(figsize=(16, 5))
    sns.countplot(data=data, x='data_ref', hue=var, palette='viridis')
    plt.title(f'Distribuição da variável "{var}" ao longo do tempo')
    plt.ylabel('Quantidade')
    plt.xlabel('Data de referência')
    plt.xticks(rotation=45)
    plt.legend(loc='upper left', title=var)
    plt.tight_layout()
    plt.show()


````

Considerações sobre os gráficos acima:

Sexo: A proporção entre homens e mulheres parece se manter relativamente constante ao longo do período analisado. Não há indicações de grandes mudanças na distribuição de gênero nos dados.

Posse de Veículo: A proporção entre aqueles que possuem e não possuem veículo se manteve aproximadamente constante ao longo do tempo.

Posse de Imóvel: A maioria das pessoas parece ter um imóvel, e essa proporção também se manteve constante ao longo do período analisado.

Tipo de Renda: Observamos algumas flutuações nas proporções entre os diferentes tipos de renda, especialmente entre "Assalariado" e "Empresário". No entanto, a distribuição parece se estabilizar nos últimos períodos.

Educação: A maioria dos indivíduos tem educação no nível "Secundário". A proporção de indivíduos com "Educação Superior" parece ter aumentado ligeiramente ao longo do tempo, enquanto a proporção de indivíduos com "Educação Secundária" diminuiu um pouco.

Estado Civil: A grande maioria dos indivíduos está classificada como "Casado". A proporção de indivíduos "Solteiros" é menor e parece ter se mantido constante ao longo do tempo.

Tipo de Residência: A vasta maioria dos indivíduos mora em "Casa", e essa proporção se manteve ao longo do período analisado.

Em resumo, a maioria das variáveis qualitativas apresentou uma distribuição relativamente constante ao longo do tempo.

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
# 1. Definindo a lista com as variáveis qualitativas e atribuindo à variável "variable"
variable = ['sexo', 'posse_de_veiculo', 'posse_de_imovel', 'tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia']

# 2. Criando a função "plot_means"
def plot_means(variables):
    for var in variables:
        plt.figure(figsize=(16, 5))
        sns.lineplot(data=data, x='data_ref', y='renda', hue=var, palette='viridis')
        plt.title(f'Média da renda por {var} ao longo do tempo')
        plt.ylabel('Média da Renda')
        plt.xlabel('Data de referência')
        plt.xticks(rotation=45)
        plt.legend(title=var, loc='upper left')
        plt.tight_layout()
        plt.show()

# 3. Chamando a função "plot_means" usando a lista "variable"
plot_means(variable)



````

Posse de Imóvel: Ter uma propriedade está associado a uma renda mais alta.

Educação: A renda tende a aumentar com o nível de educação. Por exemplo, indivíduos com educação superior geralmente ganham mais.
